```python
# Buat data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10,20,15,25,30,45,40,50,60,55,65,70]
product_B_sales = [5,10,8,15,18,20,22,30,25,35,40,45]
```

**Single Line Plot**

```python
# line plot 1 garis, style grafik default
plt.plot(months, product_A_sales, label='Produk A')
plt.title('Penjualan Produk A per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Multiple Line Plot & Customizations**

```python
# kustomisasi line plot (line 1 - product A)
plt.plot(months, product_A_sales, marker='o', linestyle='-', color='blue', label='Produk A')
# kustomisasi line plot (line 2 - product B)
plt.plot(months, product_B_sales, marker='x', linestyle='--', color='red', label='Produk B')

plt.title('Penjualan Produk per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Jenis Garis untuk Menunjukan Tren**

```python
# Data sample tambahan
product_C_sales = [15,35,20,30,45,50,45,60,55,65,50,70]
product_D_sales = [10,20,10,20,28,30,32,40,35,45,50,60]

# Plot Product A
plt.plot(months, product_A_sales, marker='o', linestyle='-', color='blue', label='Produk A')

# Plot Product B
plt.plot(months, product_B_sales, marker='x', linestyle='--', color='red', label='Produk B')

# Plot Product C
plt.plot(months, product_C_sales, marker='s', linestyle=':', color='green', label='Produk C')

# Plot Product D
plt.plot(months, product_D_sales, marker='^', linestyle='-.', color='purple', label='Produk D')

plt.title('Penjualan Produk per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

**Subplot pada Line Chart**

```python
# Subplot, 2 garis, 1 kolom, panjang 10 inch, tinggi 8 inchi
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# plot pertama untuk product C
axs[0].plot(months, product_C_sales, marker='s', linestyle=':', color='green', label='Produk C')
axs[0].set_title('Produk C')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel(' Jumlah Penjualan')
axs[0].legend()
axs[0].grid(True)


# plot pertama untuk product D
axs[1].plot(months, product_D_sales, marker='s', linestyle=':', color='purple', label='Produk D')
axs[1].set_title('Produk D')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel(' Jumlah Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
