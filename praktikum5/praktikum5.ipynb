**Praktikum 5 Visualisasi Data - Scatter Plot**


**1. Import Library**

```python
import matplotlib.pyplot as plt
```

**2. Buat Dataset**

```python
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan =[50, 60, 70, 90, 100, 110, 130, 150, 180]
```

**3. Basic Scatter Plot**

```python
plt.scatter(suhu, penjualan)
plt.title('Hubungan Penjualan Es Krim dan Suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan')
plt.show()
```

**4. Kostumisasi Scatter Plot**

```python
plt.scatter(suhu, penjualan, color='navy', s=100, edgecolor='black', alpha=0.7)
plt.title('Hubungan Penjualan Es Krim dan Suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan')
plt.grid(True)
plt.show()
```

**5. Multiple Scatter Plot**

```python
# Dataset Suhu
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]

# Dataset Penjualan
penjualan_weekdays = [50, 60, 70, 80, 90, 100, 110, 120, 130]
penjualan_weekend = [60, 70, 80, 100, 110, 120, 140, 160, 200]

plt.scatter(suhu, penjualan_weekdays, color='lime', label='Hari Kerja', s=100)
plt.scatter(suhu, penjualan_weekend, color='magenta', label='Akhir Pekan', s=100)
plt.title('Hubungan Penjualan Es Krim dan Suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan')
plt.legend()
plt.show()
```

**6. Melihat Hubungan Data dengan Beberapa Variabel**

```python
# Dataset
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan =[50, 60, 70, 90, 100, 110, 130, 150, 180]

# Dataset Tambahan
kelembapan = [60, 65, 70, 75, 80, 85, 90, 95, 100]
ukuran_titik = [200, 220, 240, 260, 280, 300, 320, 340, 360]

plt.scatter(suhu, penjualan, c=kelembapan, s=ukuran_titik, cmap='coolwarm', alpha=0.7)
plt.title('Hubungan Penjualan, Suhu, dan Kelembapan')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es Krim')
plt.colorbar(label='Kelembapan (%)')
plt.show()
```

**7. 3D Scatter Plot**

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Dataset Suhu
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]

# Dataset Penjualan
penjualan_weekdays = [50, 60, 70, 80, 90, 100, 110, 120, 130]
penjualan_weekend = [60, 70, 80, 100, 110, 120, 140, 160, 200]

# 3D Scatter Plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(suhu, penjualan_weekdays, penjualan_weekend,
           color='navy', s=300, label='Data Penjualan')

ax.set_title('Hubungan Penjualan Es Krim dan Suhu (3D Versi)')
ax.set_xlabel('Suhu')
ax.set_ylabel('Penjualan Hari Kerja')
ax.set_zlabel('Penjualan Akhir Pekan')
ax.legend()
plt.show()
```
